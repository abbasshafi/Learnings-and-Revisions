```python
#taking 5% of the total data
sample = int(0.005 * len(df))
sample
```

```python
df_sample = df.iloc[:sample,:]
df_sample.head()
```

```python
df_filtered=df_sample[df_sample["Rating"]>4.5]
df_filtered.tail()
```

```python
df_filtered.shape
```

```python
user_ratings = df_filtered.pivot_table(index='User ID', columns='Item ID', values='Rating', fill_value=0)
user_ratings.head()
```

```python
user_ratings.isnull().sum().sum()
```

```python
df
```

```python
# Similarity Calculation
user_similarity = cosine_similarity(user_ratings)
```

```python
# Example usage: Print top 5 similar users for user with
user_id = 'A0402564TCEO67AUZFJO'
user_index = user_ratings.index.get_loc(user_id)  # Get the index of the user ID
similar_users = user_similarity[user_index].argsort()[::-1][1:6]  # Exclude the given user itself and get top 5 similar users
similar_users = [user_ratings.index[user] for user in similar_users if user_ratings.index[user] in user_ratings.index]
print("Top 5 similar users for user", user_id, ":")
for user in similar_users:
    print(user)
```
